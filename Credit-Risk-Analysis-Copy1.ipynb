```python
CreditRisk_df = CreditRisk_df.drop(['R360 Reliant Score', 'Sum_Balance Per GL', 'Corp Zip'], axis=1)
CreditRisk_df
```

```python
from sklearn.preprocessing import LabelEncoder
Portfolio_dict = {'1':0,'2':0,'3':0,'4':1,'5':1,'6':1}

CreditRisk_df['Portfolio Grade'].replace(Portfolio_dict, inplace=True)

Risk_dict = {"High A Risk":0, "High B Risk":0, "High Risk":0, "Medium A Risk":1, "Medium B Risk":1, "Medium Risk":1, "Medium":1,
            "Low":2, "Low Risk":2,"Ultra-Low Risk":2 }
CreditRisk_df['Risk Tier'].replace(Risk_dict, inplace=True)

CreditRisk_df.head()
```

```python
APP_dict = {'NEW':0,'RENEW':1, 'CONC':2}

CreditRisk_df['App Type'].replace(APP_dict, inplace=True)

ISO_dict = {'WHO':0, 'DIR':1}
CreditRisk_df['ISO Group'].replace(ISO_dict, inplace=True)

CreditRisk_df.head()
```

```python
CreditRisk_df['Date_Started'] = pd.to_datetime(CreditRisk_df['Date_Started'],infer_datetime_format=True)
CreditRisk_df['Date_Started']=CreditRisk_df['Date_Started'].apply(lambda x: x.toordinal())
CreditRisk_df.dtypes
```

```python
CreditRisk_df = CreditRisk_df.dropna()
CreditRisk_df
```

```python
X = CreditRisk_df[['Avg Bank Balance','Avg Bank Negative Days','Average Net Deposits','ISO Group',
                 'Date_Started','Primary Owner FICO Score','Gross Annual Sales','Original Advance Amount',
                   'Risk Tier','Lag Days','Baseline Pace %']]
y = CreditRisk_df[['Portfolio Grade']]
```

```python
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
standar_scaler=StandardScaler()
scaled_X = standar_scaler.fit_transform(X)
```

```python
X =pd.DataFrame(scaled_X, columns=X.columns)
X.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler


X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred, normalize=True) * 100
```

```python
print("Training score:", model.score(X_train, y_train))
```

```python
parameters = {'splitter': ('best','random'),
             'max_depth': [2,4,6],
             'max_features':[2,4,6, None]}

grid_search = GridSearchCV(DecisionTreeClassifier(), parameters, cv=5,
                          return_train_score=True)
grid_search.fit(X_train, y_train)

grid_search.best_params_
```

```python
model = DecisionTreeClassifier(splitter=grid_search.best_params_['splitter'],
                                       max_depth=grid_search.best_params_['max_depth'],
                                      max_features=grid_search.best_params_['max_features'])

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
y_train_hat = model.predict(X_train)
y_pred = model.predict(X_test)

in_sample_acc = accuracy_score(y_train,y_train_hat, normalize=True) * 100

out_of_sample_acc = accuracy_score(y_test,y_pred, normalize=True) * 100

print("In-sample Accuracy:", in_sample_acc)
print("Out-of-sample Accuracy:", out_of_sample_acc)
```
